```python
# python dictionaty
people = {
    "first": ["Corey", "Jane", "John"],
    "last": ["Schefer", "Doe", "Doe"],
    "email": ["email1", "email2", "email3"]
}
import pandas as pd
df = pd.DataFrame(people)
df
```

```python
df.columns
```

```python
# rename all of the columns
df.columns = ['first_name', 'last_name', 'email']
df.columns
```

```python
# list comprehension
df.columns = [x.lower() for x in df.columns]
df
```

```python
# replace spaces with underscore instead
df.columns = df.columns.str.replace(' ', '_')
df
```

```python
# only change some column
# using rename method and pass in a dictionary of the columns that we want to change
df.rename(columns={'first_name': 'first', 'last_name': 'last'}, inplace=True)
df
```

```python
# updating the data in our rows
# update a single value
# pass in all of the new values for this row
df.loc[2] = ['John', 'Smith', 'emai3']
df
```

```python
df.loc[2, ['last', 'email']] = ['Doe', 'email3']
df
```

```python
df.loc[2, 'last'] = 'Smith'
df
```

```python
# indexer 'at': changing or looking up a single value
df.at[2, 'last'] = 'Doe'
df
```

```python
# usual mistake !error!
# !!! setting values just use .loc()

# filt = (df['email'] == 'email3')
# df[filt]['last'] = 'Smith' ❌

filt = (df['email'] == 'email3')
df.loc[filt, 'last'] = 'Smith'
df
```

```python
# update multiple rows of data
df['email'] = df['email'].str.upper()
df
```

```python
# apply: is used for calling a function on our values
# can work on either a dataframe or a series object
# apply use on a series: apply a function to every value in our series
df['email'].apply(len)
```

```python
def update_email(email):
    return email.lower()

df['email'].apply(update_email)
```

```python
df['email'] = df['email'].apply(update_email)
df
```

```python
df['email'] = df['email'].apply(lambda x: x.upper())
df
```

```python
# apply works with dataframes: runs a function on each row or column of that dataframe
# applying the function to each series in the dataframe
df.apply(len)
```

```python
len(df['email'])
```

```python
# apply to rows
df.apply(len, axis='columns')
```

```python
# grab the minimum value from each column
# series objects have a min method
df.apply(pd.Series.min)
```

```python
df.apply(lambda x: x.min())
```

```python
# applymap: apply a function to every individual element in the dataframe
# only works on dataframes
df.applymap(len)
```

```python
df.applymap(str.lower)
```

```python
# map: used for substituting each value in a series with another value 
# only works on series
# the values that we didn't substitute were converted to NaN value
df['first'].map({'Corey': 'Chris', 'Jane': 'Mary'})
```

```python
# replace
df['first'] = df['first'].replace({'Corey': 'Chris', 'Jane': 'Mary'})
df
```
